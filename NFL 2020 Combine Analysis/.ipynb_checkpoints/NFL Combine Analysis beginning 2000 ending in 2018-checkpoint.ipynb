```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("resources/nflcombine.csv", parse_dates = True)
```

### Fastest 40 time
- John Ross and Stallworth ties for a 4.22 
- Also looks as though John Ross was the fastest receiver since 2000

```python
data.sort_values("Forty").head(5)
```

### 40 Time for a RB
- Chris johnson hols the fastest forty time form 2000-2018, followed by keith marshall

```python
data[data["Pos"] == 'RB'].sort_values("Forty").head(5)
```

### Tallest Wide Receiver
- Looks like the top 5 tallest WR were all the same height

```python
data[data["Pos"] == "WR"].sort_values("Ht", ascending = False).head(5)
```

### Fastest Edge Rusher

```python
data[data["Pos"] == "EDGE"].sort_values("Forty").head()
```

### Fastest Shuttle - RB

```python
data[data["Pos"] == "RB"].sort_values("Shuttle").head()
```

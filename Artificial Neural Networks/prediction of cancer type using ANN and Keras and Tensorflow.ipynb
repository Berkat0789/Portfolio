## lets prep our data

```python
X = cancer_data.drop("target", axis=1).values
y = cancer_data["target"].values
```

## Training the data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

## Scale the data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
```

```python
X_train = scalar.fit_transform(X_train)
```

```python
X_test = scalar.transform(X_test)
```

## set up our model

```python
X_train.shape
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units = 30, activation = "relu"))
model.add(Dense(units = 15, activation = "relu"))
model.add(Dense(units = 1, activation = "sigmoid"))
model.compile(optimizer = "adam", loss = "binary_crossentropy")
```

```python
model.fit(X_train,y_train, epochs=300, validation_data=(X_test,y_test))
```

## lets see how we performed. We are overfitting so lets incoporate early stoopping to stop the model at the minal error of the val_loss 

```python
pd.DataFrame(model.history.history).plot(figsize =(16,6))
```

```python
model = Sequential()
model.add(Dense(units = 30, activation = "relu"))
model.add(Dense(units = 15, activation = "relu"))
model.add(Dense(units = 1, activation = "sigmoid"))
model.compile(optimizer = "adam", loss = "binary_crossentropy")
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(mode="min", verbose=1, patience=25)
```

```python
model.fit(X_train, y_train, epochs=500, validation_data=(X_test,y_test), callbacks=[early_stop])
```

### lets see how we performed now. Looks like we can see there is no longer overfiting and stop at the lowest erro of val loss

```python
pd.DataFrame(model.history.history)
```

```python
pd.DataFrame(model.history.history).plot(figsize = (16,6))
```

## let see if we can improve on this even more with a Dropout

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()
model.add(Dense(units = 30, activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(units = 15, activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(units = 1, activation = "sigmoid"))
model.compile(optimizer = "adam", loss = "binary_crossentropy")
```

```python
model.fit(X_train, y_train, epochs=500, validation_data=(X_test,y_test), callbacks=[early_stop])
```

## testing results

```python
pd.DataFrame(model.history.history)
```

```python
pd.DataFrame(model.history.history).plot(figsize = (16,6))
```

## Check predictions

```python
predictions = model.predict_classes(X_test)
```

```python
predictions
```

## lets check metrics

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

### Image Generator

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
image_gen = ImageDataGenerator(
    rotation_range= 20,
    width_shift_range= 0.4,
    height_shift_range= 0.4,
    shear_range= 0.4,
    zoom_range= 0.4,
    fill_mode= "nearest",
    horizontal_flip=True,
    vertical_flip= True,
    rescale=1/255
)
```

## Model generation

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()
model.add(Conv2D(activation = "relu", kernel_size = (2,2), 
                strides = (1,1), input_shape = (50,50,3), padding = "valid", filters = 32))
model.add(MaxPool2D(pool_size = (2,2)))
model.add(Conv2D(activation = "relu", kernel_size = (2,2), 
                strides = (1,1), input_shape = (50,50,3), padding = "valid", filters = 64))
model.add(MaxPool2D(pool_size = (2,2)))
model.add(Conv2D(activation = "relu", kernel_size = (2,2), 
                strides = (1,1), input_shape = (50,50,3), padding = "valid", filters = 64))
model.add(MaxPool2D(pool_size = (2,2)))
model.add(Flatten())
model.add(Dense(units = 128, activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(units = 1, activation = "sigmoid"))
model.compile(loss = "binary_crossentropy", optimizer = "adam")
```

```python
model.summary()
```

### Early Stopping

```python
stop = EarlyStopping(monitor="val_loss", mode="min", patience=2)
```

### BAtch Size 
- Many images to we need to increases the image batches

```python
batch_size = 100
```

```python
train_gen = image_gen.flow_from_directory(train_dir, target_size= (50,50), color_mode= "rgb",
                                         class_mode="binary", batch_size=batch_size,)
```

```python
test_gen = image_gen.flow_from_directory(test_dir, target_size= (50,50), color_mode= "rgb",
                                         class_mode="binary", batch_size=batch_size, shuffle=False)
```

```python
results = model.fit_generator(train_gen, validation_data=test_gen, epochs=3, callbacks=[stop])
```

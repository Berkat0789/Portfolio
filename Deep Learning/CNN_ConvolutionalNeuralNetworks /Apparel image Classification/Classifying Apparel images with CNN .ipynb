### inspect image Shape

```python
dim1 = []
dim2 = []

for image_name in os.listdir(data_dir +"/black_dress/"):
    img = imread(data_dir+"/black_dress/"+image_name)
    d1,d2,color_dim = img.shape
    dim1.append(d1)
    dim2.append(d2)
```

### Image generator

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img_gen = ImageDataGenerator(
rotation_range=0.3,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.3,
    zoom_range=0.5,
    fill_mode='nearest',
    horizontal_flip=True,
    vertical_flip=True,
    rescale=1/255,
)
```

### Testing the generator

```python
plt.imshow(img_gen.random_transform(sample_image))
```

```python
mass_shootings_2014 = pd.read_csv("GVAMassShooting2014.csv", index_col="Incident Date", parse_dates=True)
mass_shootings_2015 = pd.read_csv("GVAMassShooting2015.csv", index_col="Incident Date", parse_dates=True)
mass_shootings_2016 = pd.read_csv("GVAMassShooting2016.csv", index_col="Incident Date", parse_dates=True)
mass_shootings_2017 = pd.read_csv("GVAMassShooting2017.csv", index_col="Incident Date", parse_dates=True)
mass_shootings_2018 = pd.read_csv("GVAMassShooting2018.csv", index_col="Incident Date", parse_dates=True)
mass_shootings_2019 = pd.read_csv("GVAMassShootings2019.csv", index_col="Incident Date", parse_dates=True)
```

### Lets Analyze  2014 data for mass shootings. We will finding the incident that had the most casualties and some details on the incident.

### What state has the most mass shoots in the USAin the year 2014. What are the top 10 states

```python
mass_shootings_2014["State"].value_counts().head(1)
```

### Looks like California has the highest amount of Mass Shootings in 2014 with a count of 41. The top 10 States are located below


```python
mass_shootings_2014["State"].value_counts().head(10)
```

```python
mass_shootings_2014["State"].value_counts().head(10).plot(figsize = (16,3))
```

### Looks like the state with the most mass shooting incidents is California with 41 incident, Illinois coming in second with 26. we will evaluate the state more in dept to see which city has the most mass shooting incidents in 2014.

```python
mass_shootings_2014[mass_shootings_2014["State"] == "California"]["City Or County"].value_counts().head(5)
```

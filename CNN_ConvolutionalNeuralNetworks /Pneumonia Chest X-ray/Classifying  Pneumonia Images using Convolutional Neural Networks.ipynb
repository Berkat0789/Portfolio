### Checking image dimensions 
- Will find the average dimension of the images and use that for our image generator

```python
dim1 = []
dim2 = []
reshaped_images = []
for image_name in os.listdir(test_dir+"PNEUMONIA"):
    img = imread(test_dir+"PNEUMONiA/"+image_name)/255
    width, height = img.shape
    reshaped_images.append(img.reshape(width,height,1))
    dim1.append(width)
    dim2.append(height)
```

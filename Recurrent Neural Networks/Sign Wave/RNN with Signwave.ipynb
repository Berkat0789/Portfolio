### Train test split on this data. This will be diffferenct from the ususal train test split that we are used to. We need to decide what percentage of the end of the data set am i going to use for the testing range. So decide on what percentage of the data will be your test set 

#### 10 % 

```python
test_percent = 0.1
```

### need to determine the cutoff point

```python
len(df) * test_percent
```

### so wat what index point are we cutting this off 

```python
test_point = np.round(len(df) * test_percent)
```

```python
test_point
```

### neeed to get the test index since we have the test point. Cutoff point

```python
test_index = int(len(df) - test_point)
```

```python
test_index
```

```python
train = df.iloc[:test_index]
```

```python
train
```

```python
test = df.iloc[test_index:]
```

```python
test
```

## we will need to scale the sign column in the df. This is different then we normally do it. Because the wat RNN feed output back into them selves. We need to scale the label(sign values) -> beacuse we are passing input back into our neurons we need to scale 

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
```

```python
scalar_train = scalar.fit_transform(train)
```

```python
scalar_test = scalar.transform(test)
```

```python
### length of oupup sequnces
length  = 25
batch_size = 1
generator = TimeseriesGenerator(scalar_train, scalar_train, length=length, batch_size=batch_size)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, SimpleRNN,LSTM
```

## Number of features are we traiuning off

```python
n_features = 1
```

```python
model = Sequential()
model.add(SimpleRNN(25, input_shape = (length,n_features)))
model.add(Dense(1))
model.compile(optimizer = "adam", loss = "mse")
```

```python
model.summary()
```

```python
model.fit_generator(generator, epochs=8)
```

## model performance

```python
pd.DataFrame(model.history.history).plot()
```

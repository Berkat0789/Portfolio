```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
corona = pd.read_csv("resources/time_series_covid19_confirmed_global.csv", parse_dates = True).transpose()
```

```python
corona
```

```python
corona.drop("Province/State", inplace=True)
```

```python
corona.columns = corona.loc["Country/Region"]
```

```python
corona.drop(["Country/Region", "Lat", "Long"], inplace=True, axis=0)
```

```python
corona["US"].plot(figsize = (16,6), marker = "o")
```

## Corona shows no patterns in confirmed cases 
- Lets take a look at the growth rate daily

```python
corona["US"].pct_change().plot(figsize =(16,6))
```

```python
corona["US"].pct_change().mean() * 100
```

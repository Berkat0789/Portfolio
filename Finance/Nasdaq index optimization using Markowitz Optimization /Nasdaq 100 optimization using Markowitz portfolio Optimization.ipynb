### Random Weights

```python
number_of_ports = 2000
all_weights = np.zeros((number_of_ports, len(port_daily_returns.columns)))
all_vol = np.zeros(number_of_ports)
all_returns = np.zeros(number_of_ports)
all_sharps = np.zeros(number_of_ports)

for ind in range(number_of_ports):
    weights = np.random.random(len(port_daily_returns.columns))
    weights = weights/weights.sum()
    all_weights[ind,:] = weights
    
    all_returns[ind] = np.sum(port_daily_returns.mean() *weights) * 252 
    all_vol[ind] = np.sqrt(np.dot(weights.T, np.dot(port_daily_returns.cov() * 252, weights)))
    all_sharps[ind] = all_returns[ind]/all_vol[ind]
```

```python
max_sharp = all_sharps.argmax()
```

```python
max_return = all_returns.argmax()
```

```python
lowest_vol = all_vol.argmin()
```

```python
plt.figure(figsize=(16,8))
plt.scatter(all_vol, all_returns,c =all_sharps, alpha = 0.4)
plt.colorbar(label = "Sharp")
plt.scatter(all_vol[max_sharp], all_returns[max_sharp], c = "green", s = 50)
plt.scatter(all_vol[max_return], all_returns[max_return], c = "blue", s = 90)
plt.scatter(all_vol[lowest_vol], all_returns[lowest_vol], c="black", s = 90)
```

### Mathmatically With  Scipy

```python
from scipy.optimize import minimize
```

```python
def get_ret_vol_shp(weights):
    weights = np.array(weights)
    ret = np.sum(port_daily_returns.mean() * weights) * 252
    vol = np.sqrt(np.dot(weights.T, np.dot(port_daily_returns.cov() * 252, weights)))
    shp = ret/vol
    return np.array([ret,vol,shp])
```

```python
def check_weights(weights):
    return np.sum(weights) - 1
```

```python
def shp_check(weights):
    return get_ret_vol_shp(weights)[2] * -1
```

```python
bounds = ((0,1),) * len(passed)
```

```python
len(bounds)
```

```python
initial_guess = [0.01] * len(passed)
```

```python
len(initial_guess)
```

```python
constr = {"type":"eq", "fun": check_weights}
```

```python
best_results = minimize(shp_check, initial_guess, method="SLSQP", bounds=bounds, constraints=constr)
```

## Scipy Results 
- We get a max Sharp at 1.7 which is much higher than the 1.16 using a random allocation
- The weights for this sharp is presented below with the value of X
- We we will use this to predict the returns and vol for this portfolio

## Frontier

```python
frontier_y = np.linspace(0.19,0.35,200)
```

```python
frontier_x = []
```

```python
frontier_weights = []
```

```python
def lower_vol(weights):
    return get_ret_vol_shp(weights)[1]
```

```python
for potential_return in frontier_y:
    con = ({"type":"eq", "fun": check_weights}, {"type": "eq"
                                                 , "fun": lambda x: get_ret_vol_shp(x)[0] - potential_return })
    results = minimize(lower_vol,initial_guess, method="SLSQP", constraints=con, bounds=bounds)
    frontier_x.append(results["fun"])
    frontier_weights.append(results.x)
```

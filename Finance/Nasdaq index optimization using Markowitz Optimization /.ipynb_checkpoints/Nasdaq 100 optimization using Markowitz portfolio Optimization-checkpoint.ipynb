### Analysis - Max Return
 - We will take the 20 companies that provided that made up the nasdaq at the end of the 10 year period and put them into a portfolio
 - We will calculate the the max return, vol, and sharp for this portfolio

```python
max_return_tickers = list(random_portfolios[random_portfolios["returns"] == random_portfolios["returns"][max_return]].transpose().sort_values(286,
                                                                    ascending = False).head(13).index[3:])
```

```python
max_return_port = pd.DataFrame()
```

```python
for t in max_return_tickers:
    max_return_port[t] = web.DataReader(t, data_source = "yahoo", start = "2010-1-1")["Adj Close"]
```

```python
max_port_daily_returns = max_return_port.pct_change()
```

```python
max_return_tickers
```

```python
number_of_portfolios = 3000
all_weights = np.zeros((number_of_portfolios, len(max_return_tickers)))
all_returns = np.zeros(number_of_portfolios)
all_vol = np.zeros(number_of_portfolios)
all_sharps = np.zeros(number_of_portfolios)

for ind in range(number_of_portfolios):
    weights = np.random.random(len(max_return_tickers))
    weights = weights/weights.sum()
    all_weights[ind:,] = weights
    
    all_returns[ind] = np.sum(max_port_daily_returns.mean() * weights) * 252
    all_vol[ind] = np.sqrt(np.dot(weights.T, np.dot(max_port_daily_returns.cov() * 252, weights)))
    all_sharps[ind] = all_returns[ind]/all_vol[ind]
```

```python
max_return = all_returns.argmax()
```

```python
max_sharp = all_sharps.argmax()
```

```python
low_vol = all_vol.argmin()
```

```python
plt.figure(figsize=(16,8))
plt.scatter(all_vol,all_returns, c= all_sharps,cmap="viridis", alpha = 0.4, label = "data")
plt.scatter(all_vol[max_return],all_returns[max_return], c = "red", s = 70, label = "Max return")
plt.scatter(all_vol[max_sharp],all_returns[max_sharp], c = "blue", s = 70, label = "Max Sharp")
plt.scatter(all_vol[low_vol],all_returns[low_vol], c = "orange", s = 70, label = "Lowest Vol")
plt.colorbar(label = "Sharp");
```

```python
max_p = pd.DataFrame(all_sharps, columns=["sharp"])
```

### Scipy Calculation
- We will use to use this for validate and improve the portfolio above

```python
def get_Shrpp(weights):
    weights = np.array(weights)
    ret = np.sum(max_port_daily_returns.mean() * weights) * 252
    vol = np.sqrt(np.dot(weights.T, np.dot(max_port_daily_returns.cov() *252, weights)))
    shp = ret/vol
    return np.array([ret,vol,shp])
```

```python
def check_weights(weights):
    return np.sum(weights) - 1
```

```python
def reduce_shp(weights):
    return get_Shrpp(weights)[2] * -1
```

```python
cons = {"type":"eq", "fun": check_weights}
```

```python
bou = ((0,1),) * len(max_return_tickers)
```

```python
initial_guess = [0.01] * len(max_return_tickers)
```

```python
best_case = minimize(reduce_shp,initial_guess, method="SLSQP",constraints=cons)
```

```python
best_case
```

```python
get_Shrpp(best_case.x)
```

### Mathmatically With  Scipy -Entire S&P

```python
from scipy.optimize import minimize
```

```python
def get_ret_vol_shp(weights):
    weights = np.array(weights)
    ret = np.sum(port_daily_returns.mean() * weights) * 252
    vol = np.sqrt(np.dot(weights.T, np.dot(port_daily_returns.cov() * 252, weights)))
    shp = ret/vol
    return np.array([ret,vol,shp])
```

```python
def check_weights(weights):
    return np.sum(weights) - 1
```

```python
def shp_check(weights):
    return get_ret_vol_shp(weights)[2] * -1
```

```python
bounds = ((0,1),) * len(passed)
```

```python
len(bounds)
```

```python
initial_guess = [0.01] * len(passed)
```

```python
len(initial_guess)
```

```python
constr = {"type":"eq", "fun": check_weights}
```

```python
best_results = minimize(shp_check, initial_guess, method="SLSQP", bounds=bounds, constraints=constr)
```

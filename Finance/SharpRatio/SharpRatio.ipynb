## Let calculate the sharp ratio for the portfolio using the average daily return as well as standard deviation/Volitility of the portfolio. Annual Sharp ratio is the Risk adjusted return for the portfolio assuming risk free rate is 0 

```python
SR = (portt["Daily Return"].mean()/ portt["Daily Return"].std()) * 252**0.5
```

## We have asked the question of, how do we get the mose efficient portfolio. Where the sharp ratio is the best for our liking. Again this will all depend on an investors taste for risk. We on the other have will need to find the best weights that will satisfiy the investor liking for his or her level of risk. What we will do is take 3000 portfolios that are randomally weighted and calculate the sharp ratio. Using Markowitz effieicnt frontier we will find a few portfolios that will match to the personality of an investor. 

### lets begin with  setting up random allocations

```python
import numpy as np
```

```python
stocks  = portt[['AAPL', 'AMZN', 'IBM', 'CSCO']]
```

```python
number_port = 3000
## All random weights will go here
all_weights = np.zeros((number_port, len(stocks.columns)))
## All expected returns will go here
returns_arr = np.zeros(number_port)
## All 3000 portfolios volitilities will go here
vol_arr = np.zeros(number_port)
## All Sharp ratios will go below
sharp_array = np.zeros(number_port)

## We will loop 3000 portfolios 
for ind in range(number_port):

    ## Will need to calculate all weights that add up to 1
    wee = np.array(np.random.random(4))
    wee = wee/wee.sum()
    
    ## save weights to array
    all_weights[ind,:] = wee

    ##Calculate expected returns
    returns_arr[ind] = np.sum(por_Daily_Returns.mean() * wee *252)
    ## Calculate Portfolios Volitility 
    vol_arr[ind] = np.sqrt(np.dot(wee.T, np.dot(por_Daily_Returns.cov() * 252, wee)))
    ##Calculate the sharp ratios
    sharp_array[ind] = returns_arr[ind]/vol_arr[ind]
```

## Below we will get the max sharp ratio and its locaiton in the array. This will give us the most efficient portfolio at the desired weights

```python
sharp_array.max()
```

```python
sharp_array.argmax()
```

```python
all_weights[1658]
```

## Below are the points of the highest returns for a portfolio and lowest volitility/risk.

```python
returns_arr.argmax()
```

```python
vol_arr.argmin()
```

## We will plot all portfolios as well as the point of the highest sharp ratio, lowest Risk, and the highest returns. Agian these are the managers/investors personalities we are looking to cater to.

```python
plt.figure(figsize=(16,6))
plt.scatter(vol_arr, returns_arr, c=sharp_array, cmap="plasma")
plt.colorbar(label = "Sharp Ratio")
plt.xlabel("Volitility")
plt.ylabel("Returns")
plt.scatter(vol_arr[1658], returns_arr[1658], c = "black", s = 50, edgecolor = "red")
plt.scatter(vol_arr[2367], returns_arr[2367], c = "grey", s = 50, edgecolor = "black")
plt.scatter(vol_arr[1038], returns_arr[1038], c = "black", s = 50, edgecolor = "red")
```

### In summary the Markkowitz efficient frontier tells us that there is a certain portfolio containing certain stocks that are weighted a certain way that will be the most efficient. Meaning we will have a risk adjusted return that will give us the the highest possible return for a minimal risk. 

### So we need to consider  a few things as an investor. Do we not care about risk and will take out chances on a highly volitile portfolio but get a high return

### Do we not want much risk at all and will sacrafice a larger return to avoid a higher risk.

### Or are we in between. Willing to take on some risk for a moderate return. 

### This all depends who you are and your taste of risk is as an investor.

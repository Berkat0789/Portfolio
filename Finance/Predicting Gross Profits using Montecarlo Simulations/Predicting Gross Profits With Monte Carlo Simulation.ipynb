## Project Approach

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Company Data Simulation 
- Expected revenue for this year is 170(in millions of dollars)
- Standard deviation is 20M
- terations is 1000 - 1000 obersvations from a normal distribution

```python
rev_M = 170
stand_D = 20
iterations = 1000
```

## Simulation of future Revenues
- Expected Revenues

```python
rev = np.random.normal(rev_M, stand_D,iterations)
```

```python
plt.hist(rev, bins =50);
```

## expected revenue
- Distribution shows the mean at 170 and a std of 20M 
- We can see the expected returns values rangining from mean + 20 , and mean -20

```python
plt.plot(rev)
```

### Let scalculate COGS 
- Cost of goods Sold 
- We will assume the following 
- Cogs makes up approzimately 60% of the revenue (60% IS  used to produced to produce goods sold)
- we wilL assume COGS has an STD of 10% 
- WE will not simulate COGS 1000 times because we already have 1000 values for REV


```python
COGS = - (rev * np.random.normal(0.6,0.1))
plt.figure(figsize=(16,6))
plt.plot(COGS)
```

```python
COGS.mean()
```

```python
COGS.std()
```

### Calculating Gross Profit
- Difference between revenue and COGS

```python
gross_profit = rev + COGS
gross_profit
plt.plot(gross_profit)
```

```python
import pandas as pd
import numpy as np 
from pandas_datareader import data as web
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tickers = ["USO", "^VIX", "BND"]
data = pd.DataFrame()
for t in tickers:
    data[t] = web.DataReader(t, data_source="yahoo", start = "2007-04-10")["Adj Close"]
```

```python
fig, axes = plt.subplots(3,1, figsize = (16,12))
axes[0].plot(data["USO"], c = "r")
axes[1].plot(data["^VIX"], c = "orange")
axes[2].plot(data["BND"], c = "black")
```

### lets check the daily returns for each ticker

```python
for c in tickers:
    data[f"{c} daily returns"] = data[f"{c}"].pct_change() * 100
```
